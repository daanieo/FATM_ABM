```python
xsteps = 25

xmin = np.min(data.iloc[:,0])
xmax = np.max(data.iloc[:,0])

stepsize = (xmax-xmin)/xsteps
ymin = np.min(data.iloc[:,1])
ymax = 0
ysteps=0
ymax_real = np.max(data.iloc[:,1])
while(ymax<ymax_real):
    ymax += stepsize
    ysteps+=1
    
max_matrix = np.zeros([ysteps,xsteps])
tot_matrix = np.zeros([ysteps,xsteps])
count_matrix = np.zeros([ysteps,xsteps])
avg_matrix = np.zeros([ysteps,xsteps])
#for data x : 
# if found
# for data y :
# if found

# add to matrix 
for i in range(len(data)):
    xval = data.iloc[i,0]
    yval = data.iloc[i,1]
    
    
    for j in range(xsteps):
       
        if (xmin+j*stepsize) <= xval <= (xmin+(j+1)*stepsize):
            
            for k in range(ysteps):
                if (ymin+k*stepsize) <= yval <= (ymin+(k+1)*stepsize):
                    tot_matrix[k,j] += data.iloc[i,2]
                    count_matrix[k,j] += 1
                    avg_matrix[k,j] = tot_matrix[k,j]/count_matrix[k,j]
                    max_matrix[k,j] = np.max([max_matrix[k,j],data.iloc[i,2]])
```
