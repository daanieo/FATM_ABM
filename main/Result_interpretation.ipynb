```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

## capacitated

```python
capacity_policies =["capacitated"]#,"uncapacitated"]
access_policies  = ["base","far","tar"]
rerouting_policies = ["base","rr1","rr2"]

outcome_of_interest = "sum_uc"
# 

data_dict = {}

for cp in capacity_policies:
    for ap in access_policies:
        for rp in rerouting_policies:
            experiment_name = "%s_%s_%s"%(cp,ap,rp)
            filestring = "/home/daan/Desktop/new_run_new/%s_%s_%s_%s.csv"%(cp,ap,rp,outcome_of_interest)
            data_dict[experiment_name] = pd.read_csv(filestring,index_col='Unnamed: 0').iloc[:,1:-1]

            
```

```python
seaborn.set_context( rc={"axes.labelsize":15})
fig,ax = plt.subplots(figsize=(10,10))
runtime = 4318
x=np.linspace(0,runtime-1,4318)
for en in data_dict:
    for i in range(16):
        seaborn.lineplot(x=x,y=data_dict[en].iloc[:,1:-1].iloc[i*4])
plt.xlim([0,4318])
plt.ylim([0,0.8e5])
plt.ylabel("Unsatisfied consumption [kg]");
plt.xlabel("Cycle [10 min]");
# plt.title("Regular scenario\nUnsatisfied consumption",fontsize=15);
```

```python
m1=np.zeros([16,9])
for i in range(9):
    name = list(data_dict.keys())[i]
    data = data_dict[name]
    
    for j in range(16):
        m1[j,i]=data.iloc[j*4,-1]
list(data_dict.keys())
```

```python
m1_ordered = pd.DataFrame(m1,columns=["BB","BR1","BR2","FB","FR1","FR2","TB","TR1","TR2",]).sort_values("BB").iloc[::-1]
seaborn.set_context( rc={"axes.labelsize":15})
fig,ax = plt.subplots(figsize=(9,15))
seaborn.heatmap(data=m1_ordered,ax=ax,cmap="Blues")#,x=final_gdf_dict['fatm_id'],y=final_gdf_dict['size'],color='blue');
plt.ylabel("Run number");
plt.xlabel("Policy scenario");
# plt.title("Regular scenario\nUnsatisfied consumption",fontsize=15);
```
