```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
statistics = ["DF","ES","FS"]
alpha = []
beta = []
gamma = []
zeta = []


for a in range(6):
    alpha.append(a/10)
for b in range(6):
    beta.append(b/5)
for g in range(7):
    gamma.append(g*10/2)
for z in range(6):
    zeta.append(z/10)

alldict = {}
alldict['stat'] = list()
alldict['alpha'] = list()
alldict['beta'] = list()
alldict['gamma'] = list()
alldict['zeta'] = list()
alldict['result'] = list()


for s in statistics:
    for a in alpha:
        for b in beta:
            for g in gamma:
                for z in zeta:

                    filename = "/home/daan/GAMA/workspace/results/household_verification/nostartle/%s_alpha%s_beta%s_gamma%s_zeta%s.csv"%(s,a,b,g,z)

                    tmp= pd.read_csv(filename,skiprows=1,header=None,sep="],",engine='python') 
                                        
                    cleanlist = list()
                    for k in tmp[0][0][1:-1].split(','):

                        cleanlist.append(float(k))
                        
                        alldict['stat'].append(s)
                        alldict['alpha'].append(a)
                        alldict['beta'].append(b)
                        alldict['gamma'].append(g)
                        alldict['zeta'].append(z)
                        alldict['result'].append(k)
```

```python
dataframe = pd.DataFrame(alldict)

DF = dataframe[dataframe.stat=='DF']
results = DF.result.astype(float)
x = np.linspace(0,1440,1441)
for i in range(1512):
    plt.plot(x,results[i*1441:(i+1)*1441])
```

```python
ES = dataframe[dataframe.stat=='ES']
results = ES.result.astype(float)
x = np.linspace(0,1440,1441)
for i in range(1512):
    plt.plot(x,results[i*1441:(i+1)*1441])
```
