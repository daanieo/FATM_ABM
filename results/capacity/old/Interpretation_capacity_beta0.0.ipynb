```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.lines import Line2D
plt.style.use(["science","grid"])
```

```python
# # data1 = pd.read_feather("rerouting30_beta1.0")
# # data2 = pd.read_feather("rerouting30_beta0.0")

# # input_vars1 = pd.read_csv("input_parameters_rerouting30_beta1.0.csv")
# # input_vars2 = pd.read_csv("input_parameters_rerouting30_beta0.0.csv")

# # data2.o0 = data2.o0+(len(input_vars1))

# # data = pd.concat([data1,data2]).reset_index()
# # input_vars = pd.concat([input_vars1,input_vars2]).reset_index()

# data = pd.read_feather("capacity30_beta0.0")
# input_vars = pd.read_csv("input_parameters_capacity30_beta0.0.csv")

# data1 = pd.read_feather("./capacity/capacity30_beta0.0") # rerouting 30                        
# data2 = pd.read_feather("./capacity/capacity30_beta1.0") # rerouting 30

# input1 = pd.read_csv("./capacity/input_parameters_capacity30_beta0.0.csv")
# input2 = pd.read_csv("./capacity/input_parameters_capacity30_beta1.0.csv")

# data2.o0 = data2.o0 + len(input1)
datapol2 = pd.read_feather("capacity30_pol2")
input_varspol2 = pd.read_csv("input_parameters_capacity30_pol2.csv")

data = pd.read_feather("capacity30")#data2 #pd.concat([data1,data2]).reset_index()
input_vars = pd.read_csv("input_parameters_capacity30.csv") #pd.concat([input1,input2]).reset_index()

datapol2.o0 = datapol2.o0+len(input_vars)
input_varspol2["Unnamed: 0"] = input_varspol2["Unnamed: 0"] + len(input_vars)

for i in input_vars.index:
    if input_vars["capacity_policy"][i] == 2:
        data = data[data.o0!=i]

input_vars = input_vars[input_vars["capacity_policy"] != 2]


data = pd.concat([data,datapol2]).reset_index()
input_vars = pd.concat([input_vars,input_varspol2]).reset_index()
```

```python
def normalise(arr):
    arrmax = np.max(arr)
    arrmin = np.min(arr)
    
    arrnorm = (arr - arrmin) / (arrmax - arrmin)
    
    return arrnorm
def ticks_labels(nb_days):
    if 30%nb_days != 0:
        raise ValueError 
    else:
        vec = np.linspace(0,4320,int(30/nb_days)+1)
        labs = list()
        for i in range(int(30/nb_days)+1):
            labs.append("Day %s"%(i*nb_days))
        return vec,labs

my_palette = {}
my_palette['dark'] = ['midnightblue', 'chocolate', 'darkolivegreen']
my_palette['light'] = ['cornflowerblue','orange','yellowgreen']
my_palette['reds'] = []
```

```python
q_sum = np.zeros(np.max(input_vars["Unnamed: 0"])+1)
n_facs = 12
for j in input_vars["Unnamed: 0"]:
    tmp_sum = 0
    for i in range(n_facs):
        tmp_sum += ( np.sum(data[data.o0==j].iloc[:,i+4]))  
    
    q_sum[j]=tmp_sum    

uc_endvals_dict = {}
for i in ["unsatisfied consumption","policy type","variation","allfacs","fac5","distance"]:
    uc_endvals_dict[i] = list()
    
uc_endvals_dict["fac5"] = list()


for j in input_vars["Unnamed: 0"]:
    uc_endvals_dict["unsatisfied consumption"].append(data.o1[data.o0==j].iloc[-1])
    uc_endvals_dict["distance"].append(data.o3[data.o0==j].iloc[-1])
    uc_endvals_dict["fac5"].append( np.sum(data[data.o0==j].iloc[:,9]))
    uc_endvals_dict["allfacs"].append( q_sum[j] )

    if int(input_vars.capacity_policy[input_vars["Unnamed: 0"]==j]) > 0:
        uc_endvals_dict["policy type"].append(1)
        uc_endvals_dict["variation"].append(int(input_vars.capacity_policy[input_vars["Unnamed: 0"]==j]))
    else:
        uc_endvals_dict["policy type"].append(0)
        uc_endvals_dict["variation"].append(0)    

uc_endvals_df = pd.DataFrame(uc_endvals_dict)
```

```python
input_vars_filtered = pd.concat([input_vars,uc_endvals_df],axis=1)
```

```python

input_vars_filtered = input_vars_filtered[input_vars_filtered.beta==1]        
fig, ax = plt.subplots(2,2,figsize=(15,5))
gammas = [3,7]
scs = [6.0,7.8]

for s in range(len(scs)):    
    tmp1 = input_vars_filtered[input_vars_filtered.parallel_served_full==scs[s]]
    for g in range(len(gammas)):
        tmp2 = tmp1[tmp1.gamma==gammas[g]]
        sns.swarmplot(ax=ax[g,s] ,x="epsilon", y="unsatisfied consumption", data=tmp2,hue='capacity_policy',palette='Accent')#palette=my_palette['dark'])
        sns.violinplot(ax=ax[g,s], x="epsilon", y="unsatisfied consumption", data=tmp2,color=my_palette['light'][g])
       
        ax[g,s].set_title("gamma = %s; serving capacity = %s"%(gammas[g],scs[s]),fontsize=14)
                
        ax[g,s].set_ylabel("Unsatisfied consumption (kg)",fontsize=12)
        ax[g,s].set_xlabel(r"$\epsilon$",fontsize=12)
        handles, labels = ax[g,s].get_legend_handles_labels()
        ax[g,s].legend(handles[:4], [r"rr policy=0",r"rr policy=1",r"rr policy=2",r"rr policy=3"],loc='upper right',fontsize=12)
# plt.savefig("/home/daan/Desktop/uc_rerouting30_epsilon_full.png")
```
